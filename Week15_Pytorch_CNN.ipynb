```python
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
from glob import glob
import time

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

from torch.utils.data.sampler import SubsetRandomSampler
from torch.utils.data import Dataset

import torchvision
import torchvision.transforms as transforms
```

# Prepare Dataset

### 1. Dataset Class

```python
class InvalidDatasetException(Exception):
    def __init__(self, len_of_paths, len_of_labels):
        super().__init__(
            f"Number of paths ({len_of_paths}) is not compatible with number of labels ({len_of_labels})"
        )
```

```python
transform = transforms.Compose([transforms.ToTensor()])
```

```python
class AnimalDataset(Dataset):
    def __init__(self,img_paths,img_labels,size_of_images):
        self.img_paths = img_paths
        self.img_labels = img_labels
        self.size_of_images = size_of_images
        if len(self.img_paths) != len(self.img_labels):
            raise InvalidDatasetException(self.img_paths,self.img_labels)
        
    
    # override __len__ method
    def __len__(self):
        return len(self.img_paths)
    
    # override __getitem__ method
    # : return the image and label from index
    def __getitem__(self,index):
        PIL_IMAGE = Image.open(self.img_paths[index]).resize(self.size_of_images)
        # PIL image to Tensor
        TENSOR_IMAGE = transform(PIL_IMAGE)
        label = self.img_labels[index]
        
        return TENSOR_IMAGE, label
```

```python
paths = []
labels = []
label_map = {0:"Cat",
             1:"Dog",
             2:"Wild"
            }

for cat_path in glob("/kaggle/input/d/andrewmvd/animal-faces/afhq/train/cat/*") + glob("/kaggle/input/d/andrewmvd/animal-faces/afhq/val/cat/*"):
    paths.append(cat_path)
    labels.append(0)
    
for dog_path in glob("/kaggle/input/d/andrewmvd/animal-faces/afhq/train/dog/*") + glob("/kaggle/input/d/andrewmvd/animal-faces/afhq/val/dog/*"):
    paths.append(dog_path)
    labels.append(1)
    
for wild_path in glob("/kaggle/input/d/andrewmvd/animal-faces/afhq/train/wild/*") + glob("/kaggle/input/d/andrewmvd/animal-faces/afhq/val/wild/*"):
    paths.append(wild_path)
    labels.append(2)
    
print(len(paths))
print(len(labels))
```

```python
dataset = AnimalDataset(paths,labels,(250,250))
```

### 2. Sampler Objects

```python
from sklearn.model_selection import train_test_split

dataset_indices = list(range(0, len(dataset)))

train_indices,test_indices = train_test_split(dataset_indices,test_size=0.2,random_state=42)
print("Number of train samples: ", len(train_indices))
print("Number of test samples: ", len(test_indices))
```

```python
train_sampler = SubsetRandomSampler(train_indices)
test_sampler = SubsetRandomSampler(test_indices)
```

### 3. Data Loader Objects

```python
BATCH_SIZE = 128
train_loader = torch.utils.data.DataLoader(dataset, batch_size=BATCH_SIZE, sampler=train_sampler)
validation_loader = torch.utils.data.DataLoader(dataset, batch_size=BATCH_SIZE, sampler=test_sampler)
```

# Neural Network Modeling

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN,self).__init__()

        # define layers
        self.conv1 = nn.Conv2d(3,32,kernel_size=3,stride=2,padding=1)
        self.conv2 = nn.Conv2d(32,64,kernel_size=3,stride=2,padding=1)
        self.batchnorm1 = nn.BatchNorm2d(64)
        
        self.conv3 = nn.Conv2d(64,128,kernel_size=3,stride=2,padding=1)
        self.batchnorm2 = nn.BatchNorm2d(128)
        
        self.conv4 = nn.Conv2d(128,256,kernel_size=3,stride=2,padding=1)
        self.batchnorm3 = nn.BatchNorm2d(256)
        
        self.maxpool = nn.MaxPool2d(2,2)
        
        self.fc1 = nn.Linear(256*2*2,512)
        self.fc2 = nn.Linear(512,3)
        
    # forward
    def forward(self,x):
        # Layer 1
        x = F.relu(self.conv1(x))
        # Layer 2
        x = F.relu(self.conv2(x))
        x = self.batchnorm1(x)
        x = self.maxpool(x)
        # Layer 3
        x = F.relu(self.conv3(x))
        x = self.batchnorm2(x)
        x = self.maxpool(x)
        # Layer 4
        x = F.relu(self.conv4(x))
        x = self.batchnorm3(x)
        x = self.maxpool(x)

        # flatten
        x = x.view(-1, 256 * 2 * 2)

        # Layer 5 (fc)
        x = self.fc1(x)
        # Layer 6 (fc)
        x = self.fc2(x)
        # Layer 7
        x = F.log_softmax(x,dim=1)
        return x
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
model = CNN().to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.RMSprop(model.parameters(), lr=1e-4)
```

# Train

```python
EPOCH_NUMBER = 5
TRAIN_LOSS = []
TRAIN_ACCURACY = []

for epoch in range(1,EPOCH_NUMBER+1):
    epoch_loss = 0.0
    correct = 0
    total = 0
    for data_,target_ in train_loader:
        # one hot encoding
        target_ = target_.to(device)
        data_ = data_.to(device)
        
        # clear gradients
        optimizer.zero_grad()
        
        # loss
        outputs = model(data_)
        loss = criterion(outputs,target_)
        
        # backprop
        loss.backward()
        optimizer.step()
        
        epoch_loss = epoch_loss + loss.item()
        _,pred = torch.max(outputs,dim=1)
        correct = correct + torch.sum(pred == target_).item()
        total += target_.size(0)
    
    TRAIN_LOSS.append(epoch_loss)
    TRAIN_ACCURACY.append(100 * correct / total)
    print(f"Epoch {epoch}: Accuracy: {100 * correct/total}, Loss: {epoch_loss}")
        
```

```python
plt.subplots(figsize=(6,4))
plt.plot(range(EPOCH_NUMBER),TRAIN_LOSS,color="blue",label="Loss")
plt.legend()
plt.show()

plt.subplots(figsize=(6,4))
plt.plot(range(EPOCH_NUMBER),TRAIN_ACCURACY,color="green",label="Accuracy")
plt.legend()
plt.show()
```

# Test

```python
total_val_loss = 0.0
total_true = 0
total = len(test_sampler)

# no gradient update
with torch.no_grad():
    model.eval()
    for data_, target_ in validation_loader:
        data_ = data_.to(device)
        target_ = target_.to(device)
        
        outputs = model(data_)
        loss = criterion(outputs,target_).item()
        _,preds = torch.max(outputs, dim=1)
        total_val_loss += loss
        true = torch.sum(preds == target_).item()
        total_true += true

validation_accuracy = round(100 * total_true / total,2)
print(f"Validation accuracy: {validation_accuracy}%")
print(f"Validation loss: {round(total_val_loss, 2)}%")
```
